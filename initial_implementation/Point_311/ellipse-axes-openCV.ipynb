### Creacin de la mascara
Se usa la funcin ```create_mask_from_img``` que recibe una imagen RGBo de cualquier tipo y la convierte a una mascara binarya de valores 0 y 255.

````python
mask = pt311utils.create_mask_from_img(image)
plt.imshow(mask)
````

Esta funcin regresa un objeto tipo: RotatedRect. Este objeto se usa para aproximar una elipse.
Las propiedades que tiene son:
    *center*: This property returns the center coordinates of the rotated rectangle as a tuple (x, y).
    *size*: This property returns the dimensions of the rotated rectangle as a tuple (width, height).
    *angle*: This property returns the angle of rotation of the rectangle in degrees.

```


# Fit an ellipse to the contour points
```

````python
ellipse = pt311utils.calculate_ellipse_from_mask(mask)
````

Se usa la funcion ```draw_semi_axes_of_ellipse``` para dibujar los semi-ejes (eje desde el centro hasta la orilla) encontrados sobre la img que se le de como argumento

Se usa la funcion ```draw_axes_names``` para dibujar las letras a la mitad de cada semi-eje. Siendo la letra "a" para el semi eje mayor y la letra "b" para el semi-eje menor.

````python
pt311utils.draw_ellipse(image, ellipse)
pt311utils.draw_semi_axes_of_ellipse(image, ellipse)
pt311utils.draw_axes_names(image, ellipse)
````
